# _9.2. Dealing with Unknown Data_

```python
# unknown to nan transformation

dfHealthCare.replace(to_replace='Unknown', value = np.nan, inplace = True)
```

```python
# columns with unknown values

[(col,dfHealthCare[col].isna().sum()) for col in dfHealthCare.columns if dfHealthCare[col].isna().sum() != 0]
```

```python
# since we have many columns and the info is sensitive, we can drop the columns  >200 missing values and get a new dataframe without missing values

dfHeCaWoNa = dfHealthCare.drop(columns = ['Ntm_Speciality','Risk_Segment_During_Rx', 'Tscore_Bucket_During_Rx','Change_T_Score','Change_Risk_Segment'])
```

```python
[(col,dfHeCaWoNa[col].isna().sum()) for col in dfHeCaWoNa.columns if dfHeCaWoNa[col].isna().sum() != 0]
```

```python
# for the "Ethnicity" column, we can use mode

dfHeCaWoNa['Ethnicity'].replace(to_replace=np.nan, value = dfHeCaWoNa['Ethnicity'].mode()[0], inplace = True)
```

```python
[(col,dfHeCaWoNa[col].isna().sum()) for col in dfHeCaWoNa.columns if dfHeCaWoNa[col].isna().sum() != 0]
```

```python
# when we examine the df more, we have seen that there are also some other columns in which there are some vague values to deal

dfHeCaWoNa.replace(regex='.*Unknown.*', value = np.nan, inplace = True)
```

```python
[(col,dfHeCaWoNa[col].isna().sum()) for col in dfHeCaWoNa.columns if dfHeCaWoNa[col].isna().sum() != 0]
```

```python
# since we have many columns and the info is sensitive, we can drop the columns  >200 vague values and get a new dataframe without missing values

dfHeCaWoNa.drop(columns = ['Ntm_Speciality_Bucket'], inplace = True)
```

```python
# for other columns with vague values, we can use mode

dfHeCaWoNa['Race'].replace(to_replace=np.nan, value = dfHeCaWoNa['Race'].mode()[0], inplace = True)

dfHeCaWoNa['Region'].replace(to_replace=np.nan, value = dfHeCaWoNa['Region'].mode()[0], inplace = True)
```

```python
[(col,dfHeCaWoNa[col].isna().sum()) for col in dfHeCaWoNa.columns if dfHeCaWoNa[col].isna().sum() != 0]
```
