# _10.2b. Features to Drop Analysis_

```python
num_features = dfHealthCare.columns.to_list()

num_features.remove("Persistency_Flag")

len(num_features)
```

```python
listDrop = []

for feature in num_features:

    dfGr = dfHealthCare.groupby(by=[feature,"Persistency_Flag"])["Persistency_Flag"].value_counts().to_list()

    if (len(dfGr) == 4):
        if (dfGr[0] > dfGr[1] and dfGr[2] > dfGr[3]) or (dfGr[0] < dfGr[1] and dfGr[2] < dfGr[3]):
            listDrop.append(feature)
```
